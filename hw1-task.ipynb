```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000
```

```python
tmp = df.copy()
```

```python
def Check_LIMIT_BAL(limit_bal):
    if limit_bal <= 10000:
        return 'A'
    elif limit_bal <= 100000:
        return 'B'
    elif limit_bal <= 200000:
        return 'C'
    elif limit_bal <= 400000:
        return 'D'
    elif limit_bal <= 700000:
        return 'E'
    else:
        return 'F'

tmp['LIMIT_BAL_CLASS'] = tmp['LIMIT_BAL'].map(Check_LIMIT_BAL)
```

```python
tmp[['LIMIT_BAL_CLASS','LIMIT_BAL']].groupby('LIMIT_BAL_CLASS').agg(['count','min','max','mean','median']) # проверяем
```

```python
# (10) построить большой график (подсказка - используя seaborn) для построения завимисости всех возможных пар параметров
# разным цветом выделить разные значение "default payment next month"
# (но так как столбцов много - картинка может получиться "монструозной")
# (поэкспериментируйте над тем как построить подобное сравнение параметров)
# (подсказка - ответ может состоять из несколькольких графиков)
# (если не выйдет - программа минимум - построить один график со всеми параметрами)
import seaborn as sns
```

```python
columns = {
    'hue' : ['default.payment.next.month'],
    'base' : ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE'],
    'pays' : [x for x in df.columns if (('PAY_' in x) & ~('PAY_AMT' in x))],
    'pays_amt' : [x for x in df.columns if 'PAY_AMT' in x],
    'bills' : [x for x in df.columns if 'BILL_' in x]
}
for key, item in columns.items():
    print( '{} = {}'.format(key, item) )
```

```python
sns.pairplot( 
    df[ (columns['hue'] + columns['base']) ],
    x_vars=columns['base'],
    y_vars=columns['base'],
    hue = columns['hue'][0]
)
```

```python
sns.pairplot( 
    df[ (columns['hue'] + columns['base'] + columns['pays']) ],
    x_vars=columns['base'],
    y_vars=columns['pays'],
    hue = columns['hue'][0]
)
```

```python
sns.pairplot( 
    df[ (columns['hue'] + columns['base'] + columns['pays_amt']) ],
    x_vars=columns['base'],
    y_vars=columns['pays_amt'],
    hue = columns['hue'][0]
)
```

```python
sns.pairplot( 
    df[ (columns['hue'] + columns['base'] + columns['bills']) ],
    x_vars=columns['base'],
    y_vars=columns['bills'],
    hue = columns['hue'][0]
)
```
